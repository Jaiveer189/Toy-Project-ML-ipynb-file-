```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv('titanic.csv')
print(df.head())
```

```python
import numpy as np 
x = (1,2,3,4,5,)
y = (2,5,7,1,6)
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Plot')
plt.grid()
plt.show()
```

```python
plt.plot(x, y, color= 'blue', marker = 'x', linestyle = 'dashed')
plt.title('Sample Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```

```python
plt.hist(df['Age'].dropna(), bins = 10, color = 'blue', alpha = 0.7)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Passengers')
plt.grid(axis='x')
plt.show()
```

```python
pclass_counts = df['Pclass'].value_counts()
plt.bar(pclass_counts.index, pclass_counts.values, color='orange')
plt.xlabel('Class')
plt.ylabel('Number of Passengers')
plt.title('Passenger Class Distribution')
plt.show()
```
